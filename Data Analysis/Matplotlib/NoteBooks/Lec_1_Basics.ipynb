```python
latency = np.array([
    30,32,31,29,35,
    33,31,30,34,200
])
```

Definition (English)

Long-tail latency means a small number of requests take significantly longer than the majority of requests.

A bin is an interval that groups continuous data values together.

```python
plt.hist(latency, bins=5)
plt.xlabel("Latency")
plt.ylabel("Frequency")
plt.title("Server Latency Distribution")
plt.show()
```

```python
latency = np.random.normal(loc = 25, scale = 2, size = 1000)
```

A normal distribution is a continuous probability distribution that is symmetric around its mean, where most observations are close to the center and fewer observations appear as we move away from the mean.

```python
plt.hist(latency, bins=30, edgecolor = 'black', color = "white")
plt.xlabel("Latency")
plt.ylabel("Frequency")
plt.title("Server Latency Distribution")
plt.show()
```

```python
outliers = np.array([85, 92, 110, 88, 120, 95, 87, 105, 90, 115, 89, 99, 102, 108, 91])
```

```python
real = np.concatenate([latency, outliers])
```

```python
plt.figure(figsize=(10, 5))
plt.hist(real, bins=30, edgecolor = 'black', color = "white", alpha = 0.75, log = True)
plt.xlabel("Latency")
plt.ylabel("Frequency")
plt.title("Server Latency Distribution")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

In AI Systems and Hardware engineering, this is called Tail Latency (specifically the $p99$ or $p99.9$ latency).If you only looked at the average (mean) latency, it would barely shift because 1,000 fast runs completely mask those 15 slow ones. But looking at this histogram, you can visually spot that something is severely wrong. In a real cluster, those tiny bars represent the unfortunate users whose requests got stuck behind a background garbage collection loop, a memory page fault, or an unoptimized hardware stride.

### Why Use `log=True` for Y-Axis Scaling?

Switching from a linear to a **logarithmic scale** solves two key visualization issues:

1.  **Flattens the "Skyscraper"**  
    Compresses massive values (e.g., 1,000) so they don't dominate the chart height, freeing up visual space.
2.  **Amplifies the "Specks"**  
    Expands the lower range, turning invisible outliers (e.g., 1 or 2) into distinct, readable bars.

**Result:** A balanced view where both dominant trends and rare events are clearly visible.
