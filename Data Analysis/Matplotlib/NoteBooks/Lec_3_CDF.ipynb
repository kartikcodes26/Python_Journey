```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
latency = np.random.normal(loc = 50, scale = 1, size = 1000000)
```

```python
sorted_latency = np.sort(latency)
```

```python
cdf = np.arange(1, len(latency) + 1) / len(latency) * 100 # For percentages
```

```python
plt.plot(sorted_latency, cdf)
plt.ylabel("CDF (in %)")
plt.xlabel("Latencies (in ms)")
```
