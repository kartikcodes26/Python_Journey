```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
normal = np.concatenate([np.random.normal(loc = 50, scale = 10, size = 700), np.random.normal(loc = 100, scale = 15, size = 20)])
```

```python
optimised = np.concatenate([np.random.normal(loc = 30, scale = 10, size = 1000), np.random.normal(loc = 68, scale = 7, size = 14)])
```

```python
plt.hist(normal, color="orange", edgecolor = "black", bins=25, alpha = 0.75, log = True, label="Normal Config")
plt.hist(optimised, color="black", edgecolor = "grey", bins=25, alpha = 0.75, log = True, label="Improved Config")
plt.grid(True, linestyle='--', alpha=0.5, color="black")
plt.xlabel("Latency (Less is faster)")
plt.ylabel("Frequency")
plt.title("Server Latency Distribution")
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
```
