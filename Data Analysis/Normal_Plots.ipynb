```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = 0
low_spread = np.random.normal(mean, 5, 10000)    # Tight curve (std = 5)
norm_spread = np.random.normal(mean, 15, 10000)  # Standard curve (std = 15)
high_spread = np.random.normal(mean, 30, 10000)  # Flat curve (std = 30)
```

```python
# Plot the histograms together to compare shapes
plt.figure(figsize=(10, 5))
plt.hist(low_spread, bins=100, alpha=0.6, label='Tight (std=5)', color='black')
plt.hist(norm_spread, bins=100, alpha=0.6, label='Normal (std=15)', color='grey')
plt.hist(high_spread, bins=100, alpha=0.6, label='Flat (std=30)', color='yellow')
plt.title("How Standard Deviation Alters the Distribution Shape")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
arr = np.array([4,9,0,2,3,7,1,1])
```
