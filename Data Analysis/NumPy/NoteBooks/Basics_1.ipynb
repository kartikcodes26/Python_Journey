```python
matrix3 = np.array([[2,5,7,9],
                   [1,9,0,1],
                   [7,0,4,2]])
```

```python
print(matrix3.shape)
print(matrix3.ndim)
print(matrix3.dtype)
```

Every ndarray relies on three things to interpret that raw block of memory:

- **Data Pointer (data)**: The starting memory address of the block.
- **Shape**: The dimensions of the array (e.g., (3, 4)).
- **Strides**: A number of bytes to skip in each dimension to get to the next element. For example, jumping to the next column in a matrix takes 8 bytes, while jumping to the next row takes 32 bytes.

```python
print("Strides : ", matrix3.strides)  # prints in terms of (Row, Column)
```

```python
sub_matrix = matrix3[0:2, 0:2]
```
