```python
import numpy as np
import scipy.stats as stats
```

```python
# Attribiutes
mean = 50
std = 10
sample_size = 300
simulations = 1000
false_positives = 0
```

```python
for _ in range(simulations):
    a = np.random.normal(loc=mean, scale=10, size=sample_size)
    b = np.random.normal(loc=mean, scale=10, size=sample_size)
    _, p_value = stats.ttest_ind(a, b, equal_var=False)

    if p_value < 0.05:
        false_positives += 1
        
```

```python
print(f"False Positives : {false_positives}")
print(f"False Positives rate : {(false_positives / simulations) * 100:.2f}%")
```
