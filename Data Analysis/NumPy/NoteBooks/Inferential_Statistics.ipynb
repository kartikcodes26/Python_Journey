# Inferential Statistics
you extract a small, manageable Sample to infer the health and metrics of the entire Population.

```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
dataset = np.random.normal(loc=50, scale=10, size = 5000)
```

```python
print(f"Total mean : {np.mean(dataset)}")
print(f"Total median : {np.median(dataset)}")
```

```python
dataset_sample = np.random.choice(dataset, size = 100, replace=False)
sample_size = len(dataset_sample)
```

```python
print(f"Sample mean : {np.mean(dataset_sample)}")
print(f"Sample median : {np.median(dataset_sample)}")
```

```python
import scipy.stats as stats
```

Standard Error (SE): Think of the Standard Error exactly like standard deviation, except it measures how much our sample mean is expected to fluctuate from the true population mean.

```python
standard_error = np.std(dataset_sample) / np.sqrt(sample_size)
```

```python
z_score = 1.96 # For 95% confidence range
```

```python
margin_of_error = z_score * standard_error
```

```python
lower_bound = np.mean(dataset_sample) - margin_of_error
```

```python
upper_bound = np.mean(dataset_sample) + margin_of_error
```

```python
print(f"Sample mean : {np.mean(dataset_sample):.2f}")
print(f"Standard Error : {standard_error:.2f}")
print("I am 95% confident that the true mean lies between")
print(f"{lower_bound:.2f} to {upper_bound:.2f}")
print(f"True mean : {np.mean(dataset):.2f}")
```

# P-Value

```python
z_stastistic= np.mean(dataset) - np.mean(dataset_sample) / standard_error
```

```python
print(z_stastistic)
```

```python
p_value = stats.norm.cdf(z_stastistic)
```

```python
print(p_value)
```
