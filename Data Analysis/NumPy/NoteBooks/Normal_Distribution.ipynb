```python
import numpy as np
from IPython.display import Image, display
```

```python
np.random.seed(42)
```

```python
massive_data = np.random.normal(loc=65, scale=10, size=10000)
```

```python
print(massive_data.size)
```

```python
print(massive_data.shape)
```

```python
print(f"{np.mean(massive_data):.2f}")
```

```python
print(f"{np.median(massive_data):.2f}")
```

```python
print(f"{np.std(massive_data):.2f}")
```

# Emprirical Rule

```python
mean = np.mean(massive_data)
std = np.std(massive_data)
```

```python
step_1_low = mean - std
step_1_high = mean + std
```

```python
mask = (massive_data >= step_1_low) & (massive_data <= step_1_high)
```

```python
in_zone = massive_data[mask]
```

```python
percent = in_zone.size / massive_data.size * 100
```

```python
print(percent)
```
