# The Two Pillars of Validation

Parametric tests like ANOVA and T-tests expect your data to meet two major conditions:

## 1. Normality
The data should follow a **bell curve** (Gaussian distribution).
*   *Why it matters:* These tests assume the underlying population is normally distributed to accurately calculate probabilities.

## 2. Homoscedasticity (Homogeneity of Variance)
The groups being compared should have **roughly equal spread or variance**.
*   *Why it matters:* If one group is tightly clustered and another is wildly spread out, the test may produce false positives or lose power.

> **Note:** If your data violates these assumptions, consider using non-parametric alternatives (e.g., Kruskal-Wallis test instead of ANOVA).

# Target 1: Testing for Normality (Shapiro-Wilk Test)

Instead of visually guessing if a plot looks like a bell curve, we use the **Shapiro-Wilk test** (`scipy.stats.shapiro`) for a mathematical verdict.

## The Null Hypothesis ($H_0$)
The data **is** normally distributed.

## The Catch
This is one of the few scenarios where **you do not want a small p-value**.

| P-Value Result | Interpretation | Decision |
| :--- | :--- | :--- |
| **$p < 0.05$** | **Reject $H_0$**: The data is significantly skewed or distorted. | ❌ Do **not** proceed with parametric tests (ANOVA/t-test). |
| **$p > 0.05$** | **Fail to reject $H_0$**: The data passes the normality check. | ✅ Safe to proceed with parametric tests. |

> **Goal:** You want a **$p > 0.05$** to confirm your data follows a normal distribution.

```python
import numpy as np 
import scipy.stats as stats
```

```python
group_c = np.array([35.1, 38.4, 36.2, 34.9, 37.0, 35.5, 39.1, 36.6, 34.2, 35.8])
```

```python
shapiro_stat, shapiro_p = stats.shapiro(group_c)
```

```python
print(f"Shapiro stat : {shapiro_stat:.2f}")
print(f"Shapiro P : {shapiro_p * 100:.2f} %")
```

```python
if shapiro_p > 0.05:
    print("The Data is normally distributed")
else:
    print("The Data is not normally distributed")
```

# Target 2: Testing Equal Variance (Levene’s Test)

When comparing multiple infrastructure configurations (e.g., Groups A, B, C, and D), ANOVA assumes that while their means might differ, their background jitter (variance) is roughly identical. We use **Levene’s Test** (`scipy.stats.levene`) to verify this.

## The Null Hypothesis ($H_0$)
The variances of all groups are **equal**.

## The Rule
Just like with Normality, you want a **large p-value**.

| P-Value Result | Interpretation | Decision |
| :--- | :--- | :--- |
| **$p < 0.05$** | **Reject $H_0$**: One or more groups are significantly more volatile than others. | ❌ Breaks standard ANOVA. Use Welch’s ANOVA or a non-parametric alternative. |
| **$p > 0.05$** | **Fail to reject $H_0$**: Variances are statistically similar. | ✅ Safe to proceed with standard ANOVA. |

> **Goal:** You want **$p > 0.05$** to confirm that the "jitter" is consistent across all groups.

```python
group_A = np.array([45.2, 47.1, 44.8, 46.5, 43.9, 48.2, 45.0, 46.1, 44.3, 45.9])
group_B = np.array([44.1, 43.5, 45.2, 44.0, 43.8, 46.1, 42.9, 44.5, 45.0, 43.9])
group_C = np.array([35.1, 38.4, 36.2, 34.9, 37.0, 35.5, 39.1, 36.6, 34.2, 35.8])
group_D = np.array([44.9, 46.2, 43.8, 45.5, 47.0, 44.1, 45.3, 46.8, 43.5, 45.1])
```

```python
levene_stat, levene_p = stats.levene(group_A, group_B, group_C, group_D)
```

```python
print(f"Levene stat : {levene_stat:.2f}")
print(f"Levene P : {levene_p * 100:.2f} %")
```

```python
if levene_p > 0.05:
    print("The Variance is nearly equal")
else:
    print("The Variance's are not nearly equal")
```
