```python
import numpy as np
from IPython.display import Image, display
```

```python
scores = np.array([[30,90,81],
                   [67,52,91],
                   [48, 45, 89],
                   [78,56,85]])
```

```python
print("Global Mean : ", np.mean(scores))
print("Global Median : ", np.median(scores))
```

```python
mean_per_subject = np.mean(scores, axis=0)
```

```python
print(mean_per_subject)
```

```python
mean_per_student = np.mean(scores, axis=1)
```

```python
print(mean_per_student)
```

```python
for i, mean in enumerate(mean_per_student):
    print(f"Student {i+1} : {mean}")
```

```python
for i, mean in enumerate(mean_per_subject):
    print(f"Subject {i+1} : {mean}")
```

# Z - Scores

```python
display(Image(filename='Z-value.png'))
```

```python
arr = np.array([4,9,0,2,4,1,4,10,7])
```

```python
arr_mean = np.mean(arr)
arr_std = np.std(arr)
```

```python
z_scores = (arr - arr_mean) / arr_std
```

```python
print(z_scores)
```

```python
mask = (z_scores > 1) | (z_scores < -1)
```

```python
for i,element in enumerate(mask):
    if element:
        print(f"{arr[i]} is Outlier")
```
