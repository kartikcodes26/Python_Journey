```python
import numpy as np
import scipy.stats as stats
```

```python
mean = 50 # Null Hypothesis mean
```

```python
np.random.seed(42)
```

```python
sample = np.random.normal(loc = 46, scale = 8, size = 25)
```

```python
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # ddof=1 ensures an unbiased sample variance calculation (Bessel's correction)
```

```python
t_stats, p_value = stats.ttest_1samp(sample, popmean=mean, alternative='less')
```

```python
print(f"t_value : {t_stats:.2f}")
print(f"p_value : {p_value * 100:.4f}%")
```

Since 0.996% is way lower than your engineering threshold of 0.05 (5\%), you officially reject the Null Hypothesis. It wasn't a lucky fluke; your optimization genuinely made the system faster.Because we didn't know the true historical variance and used a small sample ($n=25$), the math automatically adjusted using the Student's t-Distribution, which has slightly thicker tails to account for that sample-size uncertainty.
