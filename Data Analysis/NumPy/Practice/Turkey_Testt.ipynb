```python
import numpy as np 
import scipy.stats as stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python

# Lower is Better
# Group A: Control / Production Baseline
group_a = np.array([45.2, 47.1, 44.8, 46.5, 43.9, 48.2, 45.0, 46.1, 44.3, 45.9])

# Group B: Customized Ring-AllReduce topology
group_b = np.array([44.1, 43.5, 45.2, 44.0, 43.8, 46.1, 42.9, 44.5, 45.0, 43.9])

# Group C: FP8 Quantized weight-loading with FlashAttention-3
group_c = np.array([35.1, 38.4, 36.2, 34.9, 37.0, 35.5, 39.1, 36.6, 34.2, 35.8])

# Group D: Hybrid Pipeline + Tensor Parallelism
group_d = np.array([44.9, 46.2, 43.8, 45.5, 47.0, 44.1, 45.3, 46.8, 43.5, 45.1])
```

```python
All_Groups = np.concatenate([group_a, group_b, group_c, group_d])
```

```python
Group_Labels = (
    ["group A"] * 10 +
    ["group B"] * 10 +
    ["group C"] * 10 +
    ["group D"] * 10
)
```

```python
f_stat, p_stat = stats.f_oneway(group_a, group_b, group_c, group_d)
```

```python
print(f"F stat : {f_stat:.2f}")
print(f"P stat : {p_stat}")
```

```python
pair_test = pairwise_tukeyhsd(endog=All_Groups, groups=Group_Labels, alpha = 0.05)
```

```python
print(pair_test)
```

```python
# C wins
```
