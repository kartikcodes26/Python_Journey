```python
import pandas as pd 
import numpy as np 
```

Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index.

```python
data = np.array([3, 9, 8.8, 8, 1])
d_index = np.array(['a', 'b', 'c', 'd', 'e'])
```

```python
s = pd.Series(data, index=d_index)
```

```python
print(s)
```

```python
print(s.index)
```

# Series can be instantiated from dicts:


```python
dict = {
    "Kartik" : 17,
    "Roshan" : 10,
    "mommy" : 40
}
```

```python
s2 = pd.Series(dict)
```

```python
print(s2)
```

```python
print(s2.index)
```

If an index is passed, the values in data corresponding to the labels in the index will be pulled out.

```python
s3 = pd.Series(dict, index = d_index)
```

```python
print(s3)
```

If data is a scalar value, the value will be repeated to match the length of index. If the index is not provided, it defaults to RangeIndex(1).

```python
s4 = pd.Series(5, index=d_index)
```

```python
print(s4)
```

Series acts very similarly to a numpy.ndarray and is a valid argument to most NumPy functions. However, operations such as slicing will also slice the index.

```python
# print(np.median(s))
mask = s > np.median(s)
print(s[mask])
```
