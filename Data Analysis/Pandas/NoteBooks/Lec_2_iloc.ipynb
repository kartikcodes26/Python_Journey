# iloc
.iloc means:

"Select by integer position."

General form:

df.iloc[rows, columns]

Think of it as

[row selection, column selection]

it also follows zero indexing

```python
df.iloc[2, 2]
```

```python
df.iloc[1:4, 2]
```

```python
df.iloc[1:4, 2:4]
```

```python
df.iloc[: , :3]

# rows: all
# columns: first three
```

```python
df.iloc[::2, ::3]

# every second item from row and every third item from column
```
