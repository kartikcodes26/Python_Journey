```python
import numpy as np
import pandas as pd

# Bigger, messier practice dataset
data = {
    "Transaction__ID": [201, 202, 203, 204, 202, 205, 206, 207, 208, 209, 203, 210, 211],
    "Customer_Name": ["john doe", "Jane Smith", "Bob Jones", "Alice Brown", "Jane Smith", "Charlie Green", "EMILY WHITE", "  Tom Clark  ", "Anna Bell", "George Best", "Bob Jones", "Ray Adams", "Sarah Jenkins"],
    "Join_Date": ["2023-01-15", "2023/02/20", "03-12-2023", "2023-04-05", "2023/02/20", "2023-06-18", "2023-07-22", None, "2023-09-01", "2023-10-10", "03-12-2023", "2023-11-15", "2023-12-25"],
    "Product_Category": ["Electronics", "Clothing", "Home", "Electronics", "Clothing", "Home", "ELECTRONICS", "Clothing", "Home", "Electronics", "Home", "Books", "Clothing"],
    "Item_Price": ["$45.00", "22.50", "$120.00", "free", "22.50", "$15.99", "$3000.00", "$12.00", None, "$85.50", "$120.00", "-5.00", "$45.00"],
    "Quantity_Ordered": [1, 2, 1, 1, 2, 3, 1, 5, 2, -4, 1, 1, 1000]
}

# Load into DataFrame
df = pd.DataFrame(data)
print("Dataset Loaded! Shape:", df.shape)
```

```python
df.head(10)
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.dropna(inplace = True)
```

```python
df['Item_Price'] = df['Item_Price'].str.replace('$', '', regex = False)
```

```python
df['Item_Price'] = df['Item_Price'].astype(float)
```

Regex is short for Regular Expression.

Think of it as a super-powered "Find and Replace" tool. While a normal search looks for exact matches (like searching for the word "free"), Regex allows you to search for patterns in text instead of exact words.

```python
df.loc[df['Item_Price'].isna(), 'Item_Price'] = 0
```

```python
df.head()
```

```python
df = df[df["Item_Price"] >= 0]
df = df[df["Quantity_Ordered"] > 0]
```

```python
df.head()
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'], format='mixed')
```
