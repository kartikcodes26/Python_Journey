```python
import numpy as np 
import scipy.stats as stats
```

# Model Problem

```python
old_model = np.array([
    245, 260, 251, 270, 238,
    290, 255, 249, 265, 272,
    258, 280, 242, 268, 256
])

new_model = np.array([
    228, 241, 233, 255, 230,
    262, 244, 235, 248, 251,
    239, 260, 226, 249, 240
])
```

```python
old_model_mean = np.mean(old_model)
new_model_mean = np.mean(new_model)
```

```python
mean = 250 # Null hypothesis
```

```python
t_value, p_value = stats.ttest_1samp(new_model, popmean=old_model_mean, alternative = 'less')
```

```python
print(f"t_value : {t_value:.2f}")
print(f"p_value : {p_value * 100:.4f}%")
```

# Battery comparison

```python
old_battery = np.array([
    10.2, 9.8, 10.5, 9.9, 10.1,
    9.7, 10.3, 10.0, 9.6, 10.4,
    9.9, 10.2
])

new_battery = np.array([
    10.8, 10.5, 11.1, 10.7, 10.9,
    10.4, 11.0, 10.6, 10.3, 10.9,
    10.7, 10.8
])
```

```python
old_battery_mean = np.mean(old_battery)
```

```python
t_valuess, p_valuess = stats.ttest_1samp(new_battery, popmean = old_battery_mean, alternative='greater')
```

```python
print(f"t_value : {t_valuess:.2f}")
print(f"p_value : {p_valuess * 100:.4f}%")
```

# Checkout Problem

```python
old_checkout = np.array([
    42.1, 41.8, 42.5, 41.9, 42.2,
    42.0, 41.7, 42.3, 42.4, 41.9
])

new_checkout = np.array([
    41.7, 41.5, 41.9, 41.6, 41.8,
    41.4, 41.6, 41.7, 41.5, 41.8
])
```

```python
print(f"t_value : {t_value:.2f}")
print(f"p_value : {p_value * 100:.4f}%")
```
