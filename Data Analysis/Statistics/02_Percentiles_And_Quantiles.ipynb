# 📦 Percentiles & Boxplots

## Percentiles

A **percentile** tells us the value below which a certain percentage of the data lies.

### Common Percentiles

| Percentile | Name | Meaning |
|------------|------|---------|
| 5th | P5 | 5% of observations are below this value |
| 25th | Q1 (First Quartile) | 25% of observations are below this value |
| 50th | Q2 (Median) | Half the observations are below this value |
| 75th | Q3 (Third Quartile) | 75% of observations are below this value |
| 95th | P95 | 95% of observations are below this value |

---

## Quartiles

Quartiles divide a sorted dataset into **4 equal parts**.

```
0% ----- Q1 ----- Median ----- Q3 ----- 100%
       25%         50%          75%
```

- **Q1:** First Quartile (25th percentile)
- **Q2:** Median (50th percentile)
- **Q3:** Third Quartile (75th percentile)

---

# Interquartile Range (IQR)

The **Interquartile Range (IQR)** measures the spread of the **middle 50%** of the data.

### Formula

```
IQR = Q3 - Q1
```

Example:

```
Q1 = 2.42
Q3 = 5.55

IQR = 5.55 - 2.42
     = 3.13
```

A larger IQR indicates more variability among the middle half of the observations.

---

# Boxplot

A **Boxplot** is a graphical summary of a dataset using quartiles.

```
           Median
             │
             ▼
──────┬───────────────┬──────
      │               │
     Q1              Q3
```

### Components

- Left edge of the box → **Q1 (25th percentile)**
- Middle line → **Median (50th percentile)**
- Right edge of the box → **Q3 (75th percentile)**
- Width of the box → **IQR**

---

# Whiskers

Whiskers extend from the box to the **smallest and largest observations that are NOT outliers**.

They **do not necessarily reach the minimum and maximum values.**

---

# Outlier Detection (1.5 × IQR Rule)

John Tukey proposed the **1.5 × IQR Rule** to identify unusually distant observations.

## Step 1

Calculate the IQR.

```
IQR = Q3 - Q1
```

## Step 2

Compute

```
1.5 × IQR
```

## Step 3

Find the fences.

### Lower Fence

```
Q1 - 1.5 × IQR
```

### Upper Fence

```
Q3 + 1.5 × IQR
```

---

## Outlier Rule

A value is considered an outlier if

```
Value < Lower Fence
```

or

```
Value > Upper Fence
```

---

## Example

Given

```
Q1 = 43
Q3 = 49
```

### Calculate IQR

```
IQR = 49 - 43
     = 6
```

### Compute

```
1.5 × IQR = 9
```

### Lower Fence

```
43 - 9 = 34
```

### Upper Fence

```
49 + 9 = 58
```

Data:

```
40 42 43 45 46 47 48 49 50 95
```

Since

```
95 > 58
```

**95 is an outlier.**

---

# Whiskers vs Fences

Do not confuse these two.

### Fences

Invisible boundaries used **only to determine outliers.**

```
Lower Fence = Q1 - 1.5 × IQR

Upper Fence = Q3 + 1.5 × IQR
```

### Whiskers

Visible lines on the boxplot that extend to the **furthest data point inside the fences.**

```
Data
|
|---- Box ----|
      │
      └── Whisker

Outside the fence → plotted as a dot (outlier)
```

---

# Reading a Boxplot

A boxplot quickly reveals:

- ✅ Median (center of the data)
- ✅ Spread of the middle 50% (IQR)
- ✅ Overall variability
- ✅ Possible skewness
- ✅ Presence of outliers

---

# Key Takeaways

- Percentiles summarize the distribution of data.
- Quartiles divide the data into four equal parts.
- Median = 50th percentile.
- IQR = Q3 − Q1 measures the spread of the middle 50%.
- A boxplot is built using Q1, Median, Q3, Whiskers, and Outliers.
- Outliers are determined using the **1.5 × IQR Rule**, not by the minimum or maximum values.
