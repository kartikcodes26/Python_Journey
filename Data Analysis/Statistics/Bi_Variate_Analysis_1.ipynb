```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

# BiVariate Analysis :

Analysing two variables at a time is called bivariate analusis

```python
# Creating a realistic dataset 

np.random.seed(42)

n = 2000

squarefeet = np.random.normal(1800, 50, n)

price = squarefeet * 220 + np.random.normal(0, 70000, n)

df = pd.DataFrame({
    "Square Feet" : squarefeet,
    "Price" : price
})

df.head()
```

```python
plt.scatter (
    df["Square Feet"],
    df["Price"],
    s = 5
)

plt.xlabel("Square Feet")
plt.ylabel("Price")
plt.title("Scatter Plot")

plt.show()
```

```python
# But for bigger data the plot becomes one big blue blob
```

# Hexbin Plot : 

```python
plt.hexbin (
    df["Square Feet"],
    df["Price"],
    gridsize = 10,
    cmap = "Blues"
)

plt.colorbar(label="Number of Houses")

plt.xlabel("Square Feet")
plt.ylabel("Price")
plt.title("Hexbin Plot")

plt.show()
```

# Contour Plot

```python
sns.kdeplot (
    x = df["Square Feet"],
    y = df["Price"],
    fill = "True"
    
)

plt.title("Contour / KDE Plot")

plt.show()
```
