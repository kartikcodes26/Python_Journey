# 📚 Chapter 2: Data and Sampling Distributions (Part 1)

## 🔄 Quick Revision

- **Population** = Entire dataset of interest.
- **Sample** = Small subset of the population.
- **Population distribution** is usually unknown; we observe the **sample distribution**.
- **Random sampling** gives every member an equal chance of being selected.
- **With replacement**: Selected item can be chosen again.
- **Without replacement**: Selected item cannot be chosen again.
- **Data quality > Data quantity** for reliable analysis.
- A good sample should be **representative** of the population.
- **Bias** = A systematic error that makes results inaccurate or unrepresentative.
- **Sample bias** occurs when the sampling method consistently favors certain groups.
- **Self-selection bias** occurs when people choose themselves to participate (e.g., online reviews).

---

## 📖 Important Terms

| Term | Meaning |
|------|---------|
| **Population** | Entire dataset you're interested in. |
| **Sample** | A subset taken from the population. |
| **N** | Population size. |
| **n** | Sample size. |
| **Random Sampling** | Every member of the population has an equal chance of being selected. |
| **Simple Random Sample (SRS)** | A sample obtained using random sampling. |
| **Bias** | A systematic error that pushes results away from the true value. |
| **Sample Bias** | When the sample does not accurately represent the population. |
| **Stratum** | A homogeneous subgroup of a population sharing common characteristics (e.g., age group, department). |
| **Stratified Sampling** | Divide the population into strata, then randomly sample from each stratum. |

---

## 💡 Key Takeaway

> **A small, random, representative sample is often more useful than a huge biased sample.**
