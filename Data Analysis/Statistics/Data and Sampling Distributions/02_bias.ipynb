## 🔄 Quick Revision

- **Random Error** causes observations to scatter randomly around the true value.
- **Bias (Systematic Error)** shifts observations consistently in one direction.
- Random error is **unavoidable**, but bias should be **minimized**.
- A model showing consistent overestimation or underestimation is likely **biased**.
- Bias often indicates:
  - Missing important variables
  - Incorrect assumptions
  - Poor sampling
  - Misspecified statistical or machine learning model
- **Random sampling** is the primary method to reduce sample bias.
- Before sampling, the **population must be clearly defined**.
- An incorrectly defined population leads to an unrepresentative sample and unreliable conclusions.

---

## 📖 Important Terms

| Term | Meaning |
|------|---------|
| **Random Error** | Natural variation that causes observations to deviate randomly from the true value. |
| **Bias (Statistical Bias)** | A systematic error that consistently shifts results away from the true value. |
| **Random Selection** | Selecting observations such that every member of the population has an equal chance of being chosen. |
| **Representative Sample** | A sample that accurately reflects the characteristics of the population. |
| **Population Definition** | Clearly specifying which individuals or observations belong to the population before sampling. |
| **Misspecified Model** | A statistical or machine learning model that produces biased results due to incorrect assumptions or missing variables. |

---

## 💡 Key Concepts

### 🎯 Random Error vs Bias

| Random Error | Bias |
|--------------|------|
| Random deviations around the true value | Consistent deviation in one direction |
| Errors tend to cancel out | Errors accumulate in the same direction |
| Difficult to eliminate completely | Can often be reduced with better methodology |

---

### 🎯 Why Random Sampling?

Random sampling helps ensure the sample is **representative** of the population by giving every member an equal probability of selection, thereby reducing sample bias.

---

### 🎯 Defining the Population

Before collecting a sample, clearly define **who belongs in the population**.

Example:
- Include active customers?
- Past customers?
- Refunded orders?
- Internal test purchases?
- Resellers?

A poorly defined population leads to poor sampling and unreliable analysis.

---

## 🤖 Data Science Insight

A machine learning model that consistently predicts values too high or too low is often **biased**.

Possible causes include:

- Missing important features
- Poor-quality training data
- Incorrect model assumptions
- Biased sampling process

---

## 💡 Key Takeaway

> **Random error creates noise, while bias creates misleading conclusions. Random sampling and a well-defined population are essential for building reliable statistical analyses and machine learning models.**

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Creating unbiased data 

np.random.seed(42)

x = np.random.normal(0, 1, 200)
y = np.random.normal(0, 1, 200)

plt.scatter(x, y)
plt.axhline(0, color = "red")
plt.axvline(0, color = "red")
plt.show()
```

```python
# Creating biased data 

np.random.seed(42)

x = np.random.normal(3, 1, 200)
y = np.random.normal(3, 1, 200)

plt.scatter(x, y)
plt.axhline(0, color = "red")
plt.axvline(0, color = "red")
plt.show()
```
