```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
population = np.random.exponential(scale = 50, size = 10000)
# Mean is 50, size is 10k 
# creates a simulated population of 100,000 exponentially distributed values with an average (mean) of 50.
```

```python
plt.figure(figsize = (12, 5))
sns.histplot(
    population,
    bins = 50, 
    kde=True
)
plt.title("Population Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
def sample_means(sample_size, n_samples = 1000):
    means = []
    for _ in range(n_samples):
        sample = np.random.choice(
            population,
            size = sample_size,
            replace = True
        )
        means.append(sample.mean())

    return means
```

```python
means_5 = sample_means(5)
sns.histplot(
    means_5,
    bins = 40,
    kde = True
)

plt.title("Sampling Distribution (n = 5)")
plt.xlabel("Sample Mean")

plt.show()
```

```python
means_5 = sample_means(20)
sns.histplot(
    means_5,
    bins = 40,
    kde = True
)

plt.title("Sampling Distribution (n = 20)")
plt.xlabel("Sample Mean")

plt.show()
```

```python
means_5 = sample_means(50)
sns.histplot(
    means_5,
    bins = 40,
    kde = True
)

plt.title("Sampling Distribution (n = 50)")
plt.xlabel("Sample Mean")

plt.show()
```
