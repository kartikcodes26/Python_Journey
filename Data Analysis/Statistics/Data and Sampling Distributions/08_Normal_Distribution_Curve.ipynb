```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

np.random.seed(42)
```

```python
data = np.random.normal(
    loc = 150, # Mean
    scale = 20, # Standard Deviation
    size = 1000 # Size
)
```

```python
sns.histplot(
    data,
    bins = 100,
    kde = True
)
plt.axvline(data.mean(), linestyle = "--", color = "red", label = "Mean")
plt.axvline(np.median(data), linestyle = "--", color = "green", label = "Median")
plt.legend()
plt.show()
```

```python
mean = data.mean()
sd = data.std()

sns.histplot(
    data,
    bins = 100,
    kde = True
)
plt.axvline(data.mean(), linestyle = "--", color = "red", label = "Mean")
plt.axvline(np.median(data), linestyle = "--", color = "green", label = "Median")
plt.axvline(mean + sd, linestyle = "--", color = "blue", label = "lowe")
plt.axvline(mean - sd, linestyle = "--", color = "blue", label = "Higher")

plt.legend()
plt.show()
```

```python
b_data = (data < mean + sd) & (data > mean - sd)
print(b_data[:20])
```

```python
print((np.sum(b_data) / 1000) * 100)
```

```python
b_data = (data < mean + 2 * sd) & (data > mean - 2 * sd)
print((np.sum(b_data) / 1000) * 100)
```
