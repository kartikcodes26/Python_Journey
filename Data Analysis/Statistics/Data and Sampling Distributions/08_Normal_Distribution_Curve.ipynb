```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

np.random.seed(42)
```

```python
data = np.random.normal(
    loc = 150, # Mean
    scale = 20, # Standard Deviation
    size = 1000 # Size
)
```

```python
sns.histplot(
    data,
    bins = 100,
    kde = True
)
plt.axvline(data.mean(), linestyle = "--", color = "red", label = "Mean")
plt.axvline(np.median(data), linestyle = "--", color = "green", label = "Median")
plt.legend()
plt.show()
```

```python
mean = data.mean()
sd = data.std()

sns.histplot(
    data,
    bins = 100,
    kde = True
)
plt.axvline(data.mean(), linestyle = "--", color = "red", label = "Mean")
plt.axvline(np.median(data), linestyle = "--", color = "green", label = "Median")
plt.axvline(mean + sd, linestyle = "--", color = "blue", label = " - 1 SD")
plt.axvline(mean - sd, linestyle = "--", color = "blue", label = " + 1 SD")

plt.legend()
plt.show()
```

```python
b_data = (data < mean + sd) & (data > mean - sd)
print(b_data[:20])
```

```python
print((np.sum(b_data) / 1000) * 100)
```

```python
b_data = (data < mean + sd) & (data > mean - sd)
print((np.sum(b_data) / 1000) * 100)

# This means 68% of data lies between +- 1 std
```

```python
b_data = (data < mean + 2 * sd) & (data > mean - 2 * sd)
print((np.sum(b_data) / 1000) * 100)

# This means 95% of data lies between +- 2 std
```

```python
b_data = (data < mean + 3 * sd) & (data > mean - 3 * sd)
print((np.sum(b_data) / 1000) * 100)

# This means 99% of data lies between +- 3 std
```

```python
sns.histplot(
    data,
    bins = 40,
    kde = True
)

for i in range (1, 4):
    plt.axvline(mean + i * sd, linestyle = "--", color = "red", label = "Higher")
    plt.axvline(mean - i * sd, linestyle = "--", color = "red", label = "lower")
    
plt.legend()
plt.title("68-95-99.7 Rule")
plt.show()

```
