# Formula Based Confidence Interval

```python
sample = np.random.normal(
    loc = 50,
    scale = 15,
    size = 1000
)
```

```python
s_mean = sample.mean()
s_std = sample.std()
n = len(sample)

SE = s_std / np.sqrt(n)

print(s_mean)
print(s_std)
print(SE)
```

```python
# For 95% confidence interval we use z = 1.96
# Because about 95% of a normal distribution lies between -1.96 and +1.96 standard deviations.
z = 1.96

margin = z * SE
print(margin)
```

```python
lower = s_mean - margin
higher = s_mean + margin
print(f"We have 95% certainity that the true mean for the entire population is between {lower} and {higher}")
```
