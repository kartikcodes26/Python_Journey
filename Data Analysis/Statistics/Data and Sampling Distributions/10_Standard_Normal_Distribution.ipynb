# Standard Normal Distribution & QQ Plot

## Standard Normal Distribution

A standard normal distribution is a normal distribution with:

- Mean = 0
- Standard Deviation = 1

It is obtained by converting every value into a **z-score**.

---

## Z-Score Formula

$$
z=\frac{x-\mu}{\sigma}
$$

Where:

- $x$ = Observation
- $\mu$ = Mean
- $\sigma$ = Standard Deviation

---

## Interpretation of Z-Score

- $z = 0$ → Exactly at the mean
- $z = 1$ → 1 standard deviation above the mean
- $z = -2$ → 2 standard deviations below the mean

Higher absolute z-score ⇒ More unusual observation.

---

## Why Standardize?

- Compare values from different datasets
- Remove units (cm, marks, salary, etc.)
- Express every value in terms of standard deviations from the mean

---

## Important Note

**Standardizing does NOT make data normally distributed.**

It only changes the **scale**:

- Mean → 0
- Standard Deviation → 1

The original shape (normal, skewed, etc.) remains unchanged.

---

## QQ Plot (Quantile-Quantile Plot)

Purpose:

Check whether sample data is approximately normally distributed.

### Interpretation

- Points follow the diagonal line → Approximately normal
- Points curve away from the line → Not normally distributed

---

## Jupyter Functions

```python
# Standardize
z = (data - data.mean()) / data.std()

# QQ Plot
from scipy import stats

stats.probplot(data, plot=plt)
plt.show()
```

---

## Key Takeaways

- Standard Normal Distribution → Mean = 0, SD = 1
- Z-score measures distance from the mean in standard deviations
- Standardization changes scale, not shape
- QQ Plot is a visual tool for checking normality

````python
data = np.random.normal(
    loc = 170,
    scale = 10,
    size = 8000
)
````

````python
z = (data - data.mean()) / data.std()
````

````python
print(z[:20])
````

````python
print(z.mean())
print(z.std())
````

````python
sns.histplot(
    z,
    bins = 100,
    kde = True
)
plt.title("Standard Normal Distribution")
plt.axvline(z.mean())
plt.show()
````

# QQ plots

````python
fig, ax = plt.subplots(figsize=(5,5))
stats.probplot(z, plot=ax)

plt.show()
````
