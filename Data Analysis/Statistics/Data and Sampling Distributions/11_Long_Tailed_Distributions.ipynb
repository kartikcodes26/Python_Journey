```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from scipy import stats

np.random.seed(42)
```

```python
long_tail = np.random.exponential(
    scale=20,
    size=10000
)

plt.figure(figsize=(8,4))
plt.hist(long_tail,bins=40,edgecolor="black")
plt.title("Long-Tailed Distribution")
plt.show()
```

```python
print(long_tail.mean())
print(np.median(long_tail))
```

```python
stats.probplot(long_tail,plot=plt)

plt.show()
```
