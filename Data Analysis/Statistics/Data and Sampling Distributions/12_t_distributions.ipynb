# Student's t-Distribution

## Why do we need it?

The normal (Z) distribution assumes the population standard deviation (σ) is known.

In reality, σ is usually unknown, so we estimate it using the sample standard deviation (s). This introduces extra uncertainty.

The t-distribution accounts for this uncertainty.

---

## Characteristics

- Similar to the normal distribution
- Has thicker (fatter) tails
- Produces wider confidence intervals
- More conservative than the normal distribution

---

## Degrees of Freedom

For a single sample:

df = n - 1

Example:
- n = 10 → df = 9
- n = 50 → df = 49

---

## Confidence Interval

Normal (Z):

x̄ ± z × (σ / √n)

Student's t:

x̄ ± t × (s / √n)

---

## Effect of Sample Size

- Small sample → t-distribution has thicker tails.
- Large sample → t-distribution becomes almost identical to the normal distribution.

---

## When to Use

Use **t-distribution** when the population standard deviation (σ) is unknown (the most common real-world case).

Use **Z-distribution** only when σ is known.

---

## Connection to Bootstrap

- t-distribution is a mathematical approximation of sampling uncertainty.
- Bootstrap estimates sampling uncertainty empirically through repeated resampling.
- Modern data science often favors bootstrap when appropriate.
