# Binomial Distribution

## Definition

The Binomial Distribution models the probability of getting exactly **k successes** in **n independent trials**, where each trial has only two possible outcomes.

---

## Conditions

A binomial experiment must satisfy:

- Fixed number of trials (n)
- Two outcomes per trial (success/failure)
- Same probability of success (p)
- Trials are independent

---

## Parameters

- n = Number of trials
- p = Probability of success

---

## Mean

μ = np

Expected number of successes.

---

## Variance

Var = np(1 − p)

Standard deviation:

SD = √(np(1 − p))

---

## Examples

- Coin tosses
- Customer purchases
- Ad clicks
- Loan defaults
- Fraud detection
- Email opens

---

## Python

```python
from scipy.stats import binom

binom.pmf(k, n, p)   # Exactly k successes
binom.cdf(k, n, p)   # At most k successes
```

---

## Relationship with Normal Distribution

For large sample sizes (large n) and when p is not too close to 0 or 1, the binomial distribution is well approximated by the normal distribution.

````python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from scipy.stats import binom

np.random.seed(42)
````

````python
binom.pmf(k = 10, n = 50, p = 0.4) # probability Exactly k successes
````

````python
binom.cdf(k = 10, n = 50, p = 0.4) # probability At most k successes
````

````python
# Graphical Example

n = 70
p = 0.2

x = np.arange(0, n - 1) 
y = binom.pmf(x, n, p)

plt.bar(
    x,
    y
)

plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n=50, p=0.5)")

plt.show()
````
