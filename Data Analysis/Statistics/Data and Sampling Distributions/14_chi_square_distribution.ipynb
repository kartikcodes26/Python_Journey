# Chi-Square Distribution

## Purpose

Measures how much observed category counts differ from expected counts.

---

## Used For

- Categorical data
- Count data
- Goodness-of-fit tests
- Independence tests
- Comparing proportions across groups

---

## Key Concepts

Observed = What actually happened.

Expected = What would happen under the null hypothesis.

---

### Key Concepts
* **Observed ($O$):** What actually happened.
* **Expected ($E$):** What would happen under the null hypothesis.

### Formula
$$\chi^2 = \sum \frac{(O - E)^2}{E}$$

---

## Properties

- Always non-negative
- Right-skewed
- Starts at 0
- Shape depends on degrees of freedom

---

## Examples

- Fair coin?
- Fair dice?
- Product preference?
- Gender vs Promotion
- Disease vs Age Group

---

## Intuition

The chi-square statistic measures **how surprising the observed counts are if the null hypothesis were true**.

```python
o = np.array([40, 60])
e = np.array([50, 50])

res = ((o - e) ** 2) / e

df = pd.DataFrame({
    "Expected": e,
    "Observed": o,
    "Chi square observation": res
})

df.head()
```
