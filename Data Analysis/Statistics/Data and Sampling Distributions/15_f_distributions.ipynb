# F-Distribution

## Purpose

Used to compare the means of multiple groups by comparing variation between groups to variation within groups.

---

## Intuition

If group means differ much more than the natural variation inside each group, the differences are likely meaningful.

---

## F-Statistic

Conceptually:

F = Between-group variation / Within-group variation

- Large F → Group means differ substantially.
- Small F → Group means are similar.

---

## Used In

- ANOVA (Analysis of Variance)
- Linear Regression

---

## Comparison

- Chi-Square → Compares category counts.
- F-Distribution → Compares continuous group means.

---

## Key Takeaway

The F-distribution helps determine whether differences among several group means are larger than would be expected from random variation alone.
