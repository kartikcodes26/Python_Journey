# Chapter 2 Revision - Data and Sampling Distributions

## Random Sampling
- Reduces bias.
- Produces representative samples.
- Larger samples generally reduce sampling error.

## Sampling Distribution
- Distribution of a statistic (e.g., sample mean) across many samples.
- Different from the distribution of raw data.

## Central Limit Theorem (CLT)
- Sample means become approximately normally distributed as sample size increases.
- Enables confidence intervals and hypothesis testing.

## Standard Error (SE)

$$
SE = \frac{s}{\sqrt{n}}
$$

- Measures variability of a sample statistic.
- Increases with larger variability.
- Decreases with larger sample sizes.

## Bootstrap
- Sample with replacement from the observed sample.
- Repeat many times.
- Estimate uncertainty without assuming a population distribution.

## Probability Distributions

### Normal Distribution
- Continuous symmetric data.
- Defined by mean (μ) and standard deviation (σ).

### Poisson Distribution
- Counts events in fixed intervals.
- Parameter: λ (average rate).

### Exponential Distribution
- Models waiting time until the next event.
- Assumes a constant event rate.

### Weibull Distribution
- Models time to failure with changing failure rates.
- Shape parameter (β):
  - β < 1 → decreasing failure rate
  - β = 1 → constant failure rate (same as Exponential)
  - β > 1 → increasing failure rate
- Scale parameter (η): characteristic lifetime.

## Key Takeaways
- Sampling introduces variability.
- Standard Error quantifies uncertainty.
- Bootstrap estimates sampling distributions.
- CLT explains why sample means are approximately normal.
- Use Normal, Poisson, Exponential, and Weibull distributions for different real-world situations.

```python
lifetimes = stats.weibull_min.rvs(
    1.5,
    scale = 5000,
    size = 100
)

lifetimes[:10]

sns.histplot(
    lifetimes,
    bins=40,
    kde = True
)
plt.title("Distribution of Simulated Weibull Lifetimes")
plt.xlabel("Lifetime (Hours)")
plt.ylabel("Frequency")
plt.show()
```
