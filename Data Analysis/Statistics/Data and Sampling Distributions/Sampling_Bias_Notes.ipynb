# 📚 Chapter 2: Selection Bias

## 🔄 Quick Revision

- **Selection Bias** occurs when data is selected or analyzed in a way that leads to misleading conclusions.
- It can happen **consciously or unconsciously**.
- Unlike **Sample Bias**, Selection Bias can occur **during both data collection and data analysis**.
- Searching through data repeatedly until something "interesting" appears is called **Data Snooping**.
- Repeatedly trying different models, features, or questions increases the chance of finding **false patterns** (Vast Search Effect).
- Always validate discoveries on **unseen (holdout/test) data**.
- **Regression to the Mean** is the natural tendency for extreme observations to be followed by more average ones.

---

## 📖 Important Terms

| Term | Meaning |
|------|---------|
| **Selection Bias** | Bias caused by selecting or analyzing data in a misleading way. |
| **Data Snooping** | Repeatedly searching through data until an interesting pattern is found. |
| **Vast Search Effect** | False discoveries caused by trying many models, variables, or analyses. |
| **Cherry Picking** | Selecting only data that supports a desired conclusion while ignoring the rest. |
| **Holdout Set** | A separate dataset used only to validate the final model and detect overfitting or false discoveries. |
| **Regression to the Mean** | The tendency of unusually high or low observations to move closer to the average in future measurements. |

---

## 🎯 Selection Bias vs Sample Bias

| Sample Bias | Selection Bias |
|--------------|----------------|
| Wrong sample is collected from the population. | Data or analysis is selected in a misleading way. |
| Happens during sampling. | Can happen during sampling **or analysis**. |
| Example: Surveying only rich people. | Trying hundreds of analyses until one gives the desired result. |

---

## 💡 Common Forms of Selection Bias

- **Non-random sampling**
- **Cherry-picking data**
- **Data snooping**
- **Vast search effect**
- **Choosing convenient time intervals**
- **Stopping an experiment early when results look interesting**

---

## 🎯 Data Snooping

> Repeatedly analyzing, filtering, or slicing data until a seemingly interesting pattern appears.

Remember:

> **"If you torture the data long enough, it will confess."**

This means that enough repeated searching can produce patterns that occur purely by chance.

---

## 🎯 Vast Search Effect

Trying many:

- Models
- Features
- Hyperparameters
- Questions

increases the probability of finding a pattern that **looks significant but is actually due to random chance**.

Always confirm such findings using a **holdout/test dataset**.

---

## 🎯 Regression to the Mean

Extreme observations are often influenced by both **skill** and **luck**.

Future observations usually contain similar skill but more average luck, causing performance to move closer to the long-term average.

Examples:

- Rookie of the Year performs less spectacularly next season.
- A student scoring 99 may score closer to their usual average on the next test.

---

## ⚠️ Don't Confuse These

| Regression to the Mean | Linear Regression |
|-------------------------|------------------|
| A statistical phenomenon | A statistical / machine learning model |
| Extreme values move toward the average | Predicts relationships between variables |

---

## 🧠 Memory Trick

### Bias = **A systematic push away from the truth.**

```
Bias
│
├── Sample Bias
│     → Wrong people selected
│
├── Statistical Bias
│     → Measurements/results systematically shifted
│
└── Selection Bias
      → Data or analysis selected in a misleading way
```

Ask yourself:

> **"Is something consistently pushing my conclusion away from the truth?"**

If **yes**, it's likely **Bias**.

---

## 💡 Key Takeaway

> **Finding an interesting pattern is easy. Proving that it is real is the difficult part.** Random sampling, proper experimentation, and validation on unseen data help distinguish genuine patterns from chance.
