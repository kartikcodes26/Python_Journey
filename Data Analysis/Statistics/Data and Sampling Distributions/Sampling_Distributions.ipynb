# 📚 Chapter 2: Sampling Distribution of a Statistic

## 🔄 Quick Revision

- Every random sample produces a slightly different statistic (e.g., mean).
- This natural variation is called **Sampling Variability**.
- A **Sampling Distribution** is the distribution of a sample statistic (such as the mean) over many random samples.
- **Data Distribution** describes the distribution of individual observations.
- **Sampling Distribution** describes the distribution of a statistic computed from many samples.
- As **sample size increases**:
  - Sample statistics become **more stable**.
  - The sampling distribution becomes **narrower**.
  - The sampling distribution becomes **more bell-shaped (approximately Normal)**.
- This behavior is explained by the **Central Limit Theorem (CLT)**.
- **Standard Error (SE)** measures how much a sample statistic varies from sample to sample.
- Larger sample sizes produce **smaller Standard Errors**, making estimates more reliable.

---

## 📖 Important Terms

| Term | Meaning |
|------|---------|
| **Sample Statistic** | A numerical measure calculated from a sample (e.g., mean, median, standard deviation). |
| **Data Distribution** | Distribution of individual observations in a dataset. |
| **Sampling Distribution** | Distribution of a sample statistic obtained from many random samples. |
| **Sampling Variability** | Natural variation in a statistic across different random samples. |
| **Central Limit Theorem (CLT)** | As sample size increases, the sampling distribution of the sample mean approaches a Normal (bell-shaped) distribution, regardless of the original data distribution (under reasonable conditions). |
| **Standard Error (SE)** | Standard deviation of a sample statistic across repeated samples; measures the precision of an estimate. |

---

## 🎯 Data Distribution vs Sampling Distribution

| Data Distribution | Sampling Distribution |
|-------------------|----------------------|
| Uses **individual observations** | Uses **sample statistics** (usually sample means) |
| Describes the dataset | Describes the variability of an estimate |
| Can be skewed, uniform, bimodal, etc. | Becomes increasingly Normal as sample size increases |
| Spread depends on the data | Spread decreases as sample size increases |

---

## 🎯 Central Limit Theorem (CLT)

The Central Limit Theorem states:

> **Regardless of the original population distribution, the sampling distribution of the sample mean becomes approximately Normal as the sample size becomes sufficiently large.**

Effects of increasing sample size:

- ✅ Sampling distribution becomes more **bell-shaped**
- ✅ Sampling distribution becomes **narrower**
- ✅ Sample means become **more consistent**

---

## 🎯 Standard Error (SE)

Measures the variability of a sample statistic across repeated samples.

### Formula

$$SE = \frac{s}{\sqrt{n}}$$

Where:

- **s** = Sample Standard Deviation
- **n** = Sample Size

---

## 🎯 Square Root Rule

To reduce the **Standard Error** by **2×**, the sample size must increase by **4×**.

Example:

```
100 samples  →  SE = X

400 samples  →  SE = X / 2
```

---

## ⚠️ Standard Deviation vs Standard Error

| Standard Deviation (SD) | Standard Error (SE) |
|--------------------------|---------------------|
| Measures variability of **individual observations** | Measures variability of a **sample statistic** |
| Describes the dataset | Describes the precision of an estimate |
| Does not necessarily decrease with larger samples | Decreases as sample size increases |
| Spread of data | Spread of sample means (or another statistic) |

## 🧠 Memory Flow

```text
Population
      │
      ▼
Random Sample
      │
      ▼
Sample Statistic (Mean)
      │
Repeat many times
      ▼
Sampling Distribution
      │
Measure its spread
      ▼
Standard Error
```

---

## 💡 Key Takeaways

- **One sample gives one estimate.**
- **Many samples reveal how reliable that estimate is.**
- **The sampling distribution studies statistics, not individual data points.**
- **Larger samples produce more stable estimates with lower Standard Error.**
- **The Central Limit Theorem explains why sample means tend to follow a Normal distribution, even when the original data does not.**
