```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population = [
    563626, 710231, 1200000, 1500000, 1800000,
    2200000, 2600000, 3000000, 3500000, 4000000,
    4500000, 5200000, 6000000, 7000000, 8000000,
    10000000, 12000000, 15000000, 20000000, 37253956
]
```

```python
df = pd.DataFrame({
    "population" : population
})
```

```python
print(df)
```

```python
# Separating values into groups of ranges
bins = pd.cut(df['population'], bins = 10)
```

```python
print(bins)
```

```python
frequency = bins.value_counts()
print(frequency)
```

```python
# Plotting a histogram

df['population'].plot.hist(
    bins = 20,
    edgecolor = "Black"
)

plt.xlabel("Population")
plt.ylabel("Count")
plt.title("Population")

plt.show()
```

```python
df['population'].plot.density()

plt.xlabel("Population")
plt.ylabel("Density")
plt.title("Population")

plt.show()
```

```python
ax = df['population'].plot.hist(
    bins = 20,
    edgecolor = "Black",
    density = True
    )

df['population'].plot.density(ax=ax)

plt.xlabel("Population")
plt.ylabel("Histogram with density curve")
plt.title("Population")

plt.show()
```

```python
df2 = pd.DataFrame({"Frequency" : bins.value_counts().sort_index()})
```

```python
print(df2)
```

```python
# Box Plot
df["population"].plot.box()

plt.ylabel("Population")

plt.show()
```
