```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

# Set seed for reproducibility
np.random.seed(42)
```

```python
sns.get_dataset_names()
```

```python
plt.style.use("dark_background")

# Apply Seaborn theme while preserving Matplotlib's dark axes/figure parameters
sns.set_theme(
    style="darkgrid", 
    rc={
        "axes.facecolor": "#121212",
        "figure.facecolor": "#121212",
        "grid.color": "#333333",
        "text.color": "white",
        "axes.labelcolor": "white",
        "xtick.color": "white",
        "ytick.color": "white",
    }
)
```

```python
df = sns.load_dataset("penguins")
```

```python

sns.scatterplot(
    data = df,
    x = "bill_length_mm",
    y = "bill_depth_mm",
    hue = "sex"
)

plt.show()
```

```python
sns.despine() # Remove the top and the right spines

sns.scatterplot(
    data = df,
    x = "bill_length_mm",
    y = "bill_depth_mm",
    hue = "sex"
)

plt.show()
```

```python
sns.set_context("paper") # paper, notebook, talk, poster

sns.scatterplot(
    data = df,
    x = "bill_length_mm",
    y = "bill_depth_mm",
    hue = "sex"
)

plt.show()
```

```python
sns.stripplot(
    data = df,
    x = "flipper_length_mm",
    y = "body_mass_g",
    hue = "sex"
)

plt.show()
```

```python
sns.stripplot(
    data = df,
    x = "species",
    y = "body_mass_g",
    hue = "island",
    jitter=True,
    dodge = True,
)

plt.show()
```

```python
sns.swarmplot(
    data = df,
    x = "species",
    y = "body_mass_g",
    hue = "island",
    dodge = True,
)
```

# Histogram

```python
sns.histplot(
    data = df,
    x = "body_mass_g",
    hue = "sex",
    multiple = "stack"
)

plt.show()
```

```python
# regression plot 
sns.regplot(
    data = df,
    x = "body_mass_g",
    y = "flipper_length_mm",
    scatter = False
)

plt.show()
```

```python
# Line Plot

sns.lineplot(
    data = df,
    x = "body_mass_g",
    y = "flipper_length_mm",
    hue = "sex"
)

plt.show()
```

```python
# bar plot 
sns.barplot(
    data = df,
    x = "island",
    y = "flipper_length_mm",
    hue = "sex",
    estimator = np.std
)

plt.show()
```

```python
# Box Plot

sns.boxplot(
    data = df,
    x = "island",
    y = "flipper_length_mm",
    hue = "sex",
)

plt.show()
```

```python
# violin Plot

sns.violinplot(
    data = df,
    x = "island",
    y = "flipper_length_mm",
    hue = "sex",
    split = True
)

plt.show()
```

```python
sns.kdeplot(
    data = df,
    x = "flipper_length_mm",
    hue = "sex",
    fill = True
)

plt.show()
```
