# Expected Value

Expected value is the sum of values times their probability of occurrence, often
used to sum up factor variable levels

EX = Cost * Probability

The probabiliy must sum to zero

```python
df = pd.DataFrame({
    "Revenue": [500, 300, 100, 0],
    "Probability": [0.2, 0.3, 0.2, 0.3]
})
```

```python
df["Weight"] = df["Revenue"] * df["Probability"]
```

```python
df.head()
```

```python
expected_value = df["Weight"].sum()
print(expected_value)
```
