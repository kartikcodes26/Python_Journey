## A/B Testing

### Definition
- Compares **two treatments (A & B)** to determine which performs better.

### Key Terms
- **Treatment:** The change being tested (price, UI, drug, headline, etc.).
- **Subject:** Entity receiving the treatment (user, patient, seed, etc.).
- **Treatment Group:** Receives the new treatment.
- **Control Group:** Receives the existing/no treatment (baseline).
- **Randomization:** Randomly assigns subjects to groups to reduce bias.
- **Test Statistic (Metric):** Measure used to compare groups (CTR, conversion rate, revenue, watch time, etc.).

### Why Randomization?
Ensures differences between groups are due to:
1. Actual treatment effect.
2. Random chance.

### Why a Control Group?
Without a control group, external factors (season, weather, marketing, etc.) may influence results, making it impossible to isolate the treatment's effect.

### Types of Metrics
- **Binary:** Click/No Click, Buy/Don't Buy, Fraud/Not Fraud.
- **Continuous:** Revenue, Purchase Amount, Time Spent, Pages Visited.

### Blinding
- **Blind Study:** Subjects don't know their treatment.
- **Double-Blind Study:** Neither subjects nor researchers know.

### Important Notes
- Decide the **primary metric before** running the experiment.
- Don't blindly trust every statistic software outputs.
- A/B testing can be extended to multiple treatments (A/B/C/D...), leading to **Multi-Arm Bandit** experiments.

### Key Takeaways
- Random assignment minimizes bias.
- Control group provides a fair baseline.
- Measure success using a predefined metric.
- Differences may arise from either a real treatment effect or random variation.
