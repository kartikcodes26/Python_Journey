## Hypothesis Tests

### Purpose
- Determines whether an observed difference is **real** or simply due to **random chance**.
- Protects us from seeing patterns where none actually exist.

---

### Why Do We Need It?
After an A/B test, any observed difference can be due to:
1. **Random Chance**
2. **A Real Treatment Effect**

Hypothesis testing helps determine which is more likely.

---

### Null Hypothesis (H₀) & Alternative Hypothesis (H₁ / H␡)

<div align="center">
  <img src="https://images.unsplash.com/photo-1607604276583-eef5d076aa5f?auto=format&fit=crop&w=800&q=80" alt="Decision Making and Hypotheses" width="600" />
</div>

#### Null Hypothesis (H₀)
- Assumes **no real difference** exists.
- Any observed difference is due to chance.
- Default assumption before analyzing data.

**Examples:**
- New website = Old website
- New medicine is no better than the old one.

#### Alternative Hypothesis (H₁ / H␡)
- Opposite of the null hypothesis.
- Represents what you're trying to prove.

**Examples:**
- New website ≠ Old website
- New medicine is better than the old one.

---

### Null Model
- A probability model built assuming **H₀ is true**.
- Used to check whether the observed result is unusually rare.

---

### One-Tailed vs. Two-Tailed Tests

<div align="center">
  <img src="https://images.unsplash.com/photo-1543286386-713bdd548da4?auto=format&fit=crop&w=800&q=80" alt="Directional Statistical Testing" width="600" />
</div>

#### One-Tailed Test
Use when **only one direction matters**.

**Example:**
- Is the new website **better**?
- Is the new price **more profitable**?

*(If it's worse, simply keep the old version.)*

#### Two-Tailed Test
Use when **both directions matter**.

**Example:**
- Is the new medicine **different** (better or worse)?
- Are the average salaries different?

---

### Easy Rule
- **Only care if it's better?** → One-tailed test.
- **Care if it's different (better or worse)?** → Two-tailed test.

---

### Key Takeaways
- Always begin by assuming **H₀ is true**.
- We **reject H₀**, not "prove H₁".
- One-tailed tests check one direction.
- Two-tailed tests check both directions.
- Hypothesis testing is about distinguishing **signal** from **random noise**.
