```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 

np.random.seed(42)
```

```python
page_a = np.array([12, 15, 18, 20, 22])
page_b = np.array([25, 27, 30, 32, 35])
```

```python
diff = page_b.mean() - page_a.mean()
print(diff)
```

```python
combined = np.concatenate([page_a, page_b])
print(combined)
```

```python
np.random.shuffle(combined) # shuffle in place
print(combined)
```

```python
r_means = []
for _ in range(1000):
    np.random.shuffle(combined)
    a = combined[:5]
    b = combined[5:]
    r_means.append(b.mean() - a.mean())

print(r_means[:10])
```

```python
sns.histplot(
    data = r_means,
    bins = "auto",
    kde=True
)
plt.axvline(diff, linestyle = "--", color = "red")
plt.show()
```
